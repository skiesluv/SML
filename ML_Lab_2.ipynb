```python
import random
import numpy as np
import unittest
import time

def makeRandomMatrixNotInt(n,m):
  return [[random.random() for row in range(n)] for column in range(m)]

def makeZeroMatrix(n,m):
  return [[random.randint(0, 0) for row in range(n)] for column in range(m)]
```

```python
m1r = makeRandomMatrixNotInt(25,25)
m2r = makeRandomMatrixNotInt(25,25)

print("Matrix №1 = ", m1r)
print("Matrix №2 = ", m2r)

m3r = makeZeroMatrix(25,25)

print("Zero Matrix (№3) = ", m3r)

m3result = makeZeroMatrix(25,25)

print("Zero Matrix (№4) = ", m3result)
```

```python
start_time = time.time()

for i in range(25):
  for j in range(25):
    sum = 0
    for k in range(25):
      sum += (m1r[i][k] * m2r[k][j])
    m3result[i][j] = sum

print("The multiplication of Matrix №1 and №2")
print(m3result)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
def matrixmult(n,m1r,m2r,m3result):
  for i in range(n):
    for j in range(n):
      sum = 0
      for k in range(n):
        sum += (m1r[i][k] * m2r[k][j])
      m3result[i][j] = sum
  return isinstance(m3result, list)

def matrixvectormult(N, m3result, vectorN):
  c = [0]*(N)
  for i in range(N):
    c[i]=0

  for i in range(N):
    for j in range(N):
      c[i] += (m3result[i][j] * vectorN[j])
  return isinstance(c, list)

def vectormatrixmul(N, m3result, vectorN):
  b = [0]*(N)
  for i in range(N):
    b[i]=0

  for i in range(N):
    for j in range(N):
      b[i] += (vectorN[j] * m3result[j][i])
  return isinstance(b, list)

def vectormulvector(N, vectorN, vectorX, vectorZeroN):
  x = 0
  for i in range(N):
    vectorZeroN[i] = vectorN[i] * vectorX[i]
    x += vectorZeroN[i]
  return isinstance(x, float)
```

```python
class MatMulTest(unittest.TestCase):
    def test_mat_mul(self):
        self.assertTrue(matrixmult(25,m1r,m2r,m3result))
if __name__ == "__main__":
  unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```python
class MatMulVectorTest(unittest.TestCase):
  def test_mat_vector_mul(self):
    self.assertTrue(matrixvectormult(25, m3r, vectorN))
if __name__ == "__main__":
  unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
