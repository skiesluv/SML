```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

#Створення набору даних
X, y = make_blobs(
   n_samples=100, n_features=2,
   centers=4, cluster_std=0.5,
   shuffle=True, random_state=0
)

#Графік набору даних
plt.scatter(
   X[:, 0], X[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
```

```python
#Створення 4 кластерів за допомогою KMeans
from sklearn.cluster import KMeans

km = KMeans(
    n_clusters=4, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(X)
```

```python
#Будування графіку для 4 кластерів
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='#48fc00',
    marker='s', edgecolor='black',
    label='Cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='#fc6900',
    marker='o', edgecolor='black',
    label='Cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='#00b9fc',
    marker='v', edgecolor='black',
    label='Cluster 3'
)

plt.scatter(
    X[y_km == 3, 0], X[y_km == 3, 1],
    s=50, c='#bf23de',
    marker='p', edgecolor='black',
    label='Cluster 4'
)

#Відображення центроїд для кожного з кластерів
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='Centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
#Cтатистичний аналіз для кожного сформованого кластеру
import statistics
print('Cluster 1 (Green) : Median =', (statistics.median(X[y_km == 0, 1])), '| Mean =', (statistics.mean(X[y_km == 0, 1])), '| Deviation =', (statistics.stdev(X[y_km == 0, 1])))
print('Cluster 2 (Orange) : Median =', (statistics.median(X[y_km == 1, 1])), '| Mean =', (statistics.mean(X[y_km == 1, 1])), '| Deviation =', (statistics.stdev(X[y_km == 1, 1])))
print('Cluster 3 (Blue) : Median =', (statistics.median(X[y_km == 2, 1])), '| Mean =', (statistics.mean(X[y_km == 2, 1])), '| Deviation =', (statistics.stdev(X[y_km == 2, 1])))
print('Cluster 4 (Purple) : Median =', (statistics.median(X[y_km == 3, 1])), '| Mean =', (statistics.mean(X[y_km == 3, 1])), '| Deviation =', (statistics.stdev(X[y_km == 3, 1])))
```
